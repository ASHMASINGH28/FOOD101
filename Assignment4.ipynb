```python
#importing necessary libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LeakyReLU, BatchNormalization, Dropout
```

```python
df= pd.read_csv("updated_pollution_dataset.csv")
df
```

```python
# Obtaining X and Y (Labels) from the dataframe.
X = df.drop(columns=['Air Quality'])
Y = df['Air Quality']

#Splitting the data into training and testing subsets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# One hot encoding of categorical data
unique_categories , indices = tf.unique(Y_train)
num_categories = len(unique_categories)
Y_train= tf.one_hot(indices, depth=num_categories)

uniquecategories , indices = tf.unique(Y_test)
numcategories = len(uniquecategories)
Y_test= tf.one_hot(indices, depth=numcategories)

```

```python
# Standardize the input data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Define the neural network architecture (Using simple MLP)
model = Sequential()
model.add(Dense(625, input_dim= 9))
model.add(LeakyReLU(alpha=0.1))
model.add(BatchNormalization())
model.add(Dense(512))
model.add(LeakyReLU(alpha=0.1))
model.add(BatchNormalization())
model.add(Dense(225))
model.add(LeakyReLU(alpha=0.1))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(20))
model.add(Dense(4, activation='softmax'))



# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Training
history = model.fit(X_train, Y_train, epochs=50, batch_size=50, validation_data=(X_test, Y_test))

# Evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, Y_test)
print(f'Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}')
```
